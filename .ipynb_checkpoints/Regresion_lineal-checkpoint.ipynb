# Regresion Lineal

```python
import matplotlib.pyplot as plt
# Dataset
head = [4512,3738,4261,3777,4177,3585,3785,3613,3982,3443,
        3993,3640,4208,3832,3876,3497,3466,3095,4424,3878,
        4046,3804,3710,4423,4036,4022,3454,4175]
brain = [1530,1297,1335,1282,1590,1300,1400,1255,1355,1375,
         1340,1315,1522,1208,1405,1358,1292,1340,1400,1357,
         1287,1275,1270,1505,1490,1485,1310,1420]
# Graficar mis datos
# plt.plot(head,brain,'ro')
# plt.show()
# Regresion lineal
# Valores iniciales
w0 = 585
w1 = 0.3
x0 = 1
x1 = head
y = brain
k = 1.0E-9
# Numero de iteracion
n_iteracion = 0
# variables para actualizar pesos
w_viejo = w0
w_nuevo = 0
# Control de iteraciones
for i in range(100):
    # Calcular f(x)=x0*w0+x1*w1
    fx = [x0*w0+x1[i]*w1 for i in range(len(head))]
    # Calcular y-f(x)
    yfx = [y[i]-fx[i] for i in range(len(head))]
    # Calcular (-x0)(y-f(x))
    x0yfx = [-x0*yfx[i] for i in range(len(head))]
    # Calcular la sumatoria de S((-x0)(y-f(x)))
    Sx0 = 0
    for i in range(len(head)):
        Sx0 += -x0*yfx[i]
    # Calcular (-x1)(y-f(x))
    x1yfx = [(-x1[i])*yfx[i] for i in range(len(head))]
    Sx1 = 0
    for i in range(len(head)):
        Sx1 += (-x1[i])*yfx[i]
    # Actualizacion de pesos
    print("Iteracion #" + str(n_iteracion) + " => \tw0 = " + str(w0) +"\t\tw1 = " + str(w1))
    w0 = w0-(k*Sx0)
    w1 = w1-(k*Sx1)
    n_iteracion += 1
    
# Graficar mis datos
x = [i for i in range(6000)]
y = [w1*x+w0 for x in range(6000)]
h1 = [w1*x+w0+200 for x in range(6000)]
h2 = [w1*x+w0-200 for x in range(6000)]
plt.plot(head, brain, 'ro', x, y, '', x, h1, '--', x, h2, '--')
plt.show()
```
